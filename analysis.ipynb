```python
import pandas as pd
import numpy as np

ibm = pd.read_csv('outputs/ibm-2022-apriori.csv')
ibm
```

```python
def generate_source_target_col(df):
    df['ant_count'] = [len(s.split(' '))+1 for s in df.antecedent]
    df['con_count'] = [len(s.split(' '))+1 for s in df.consequent]
    return df


def analysis(df):
    df = generate_source_target_col(df)
    # get support, confidence threshold
    quantile = df[['support', 'confidence']].quantile(np.arange(0, 1, 0.1))
    low_support = quantile.loc[.2, 'support']
    high_support = quantile.loc[.8, 'support']
    low_confidence = quantile.loc[.2, 'confidence']
    high_confidence = quantile.loc[.8, 'confidence']
    print('quantiled support and confidence')
    display(quantile)

    ll = df[(df.support < low_support) & (df.confidence < low_confidence)]
    lh = df[(df.support < low_support) & (df.confidence > high_confidence)]
    hl = df[(df.support > high_support) & (df.confidence < low_confidence)]
    hh = df[(df.support > high_support) & (df.confidence > high_confidence)]

    print('low_support, low_confidence')
    display(ll.describe())

    print('low_support, high_confidence')
    display(lh.describe())

    print('high_support, low_confidence')
    display(hl.describe())

    print('high_support, high_confidence')
    display(hh.describe())
analysis(ibm)

# ibm.quantile([.1, .9])
```
